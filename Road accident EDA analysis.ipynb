# Exploratory Data Analysis

```python
df.columns
```

```python
#counting vaules of roadtype and renaming axis 

df3= df['Road_Type'].value_counts().rename_axis('Road_Type').reset_index(name='counts')
df3
```

```python
# assigning above table values to the variables

y_lables= df3.Road_Type
y_values= df3.counts
y_explode=(0.1,0,0,0,0.3)
```

```python
# using matplotlib to plot pie

plt.pie(y_values, labels=y_lables, autopct='%1.2f%%',explode=y_explode)
plt.title('Road_Type')
plt.show()
```

* The pie chat represent percentage of Road Type WRT Accident caused.

* Major Accidents accured in the Single carriageway

* Least @ slip road as per the dataset

```python
# plotting countplot using seaborn

plt.figure(figsize=(5,5))
ax=sns.countplot(data=df,y='Road_Surface_Conditions')
for bars in ax.containers:
       ax.bar_label(bars)
```

* This represents the total counts of Road surface conditions accidents
* Dry surface counts witnesses major accident surface condition

```python
# plotting countplot using seaborn

plt.figure(figsize=(5, 5))
ax= sns.countplot(x='Accident_Severity',data= df)
for bars in ax.containers:
        ax.bar_label(bars)
```

* Represents the Accident Severity 
* High no. of cases are filled for slight category
* Low no. of cases filled for fatal category

```python
# plotting data groupby using seaborn

weather_vehicle = df.groupby(['Weather_Conditions'], as_index=False)['Number_of_Vehicles'].sum().sort_values(by='Number_of_Vehicles', ascending=True).head()

sns.set(rc={'figure.figsize':(10,5)})
ax=sns.barplot(data = weather_vehicle, x = 'Weather_Conditions',y= 'Number_of_Vehicles')
for bars in ax.containers:
    ax.bar_label(bars)
```

* Represents the number of vehicles met accident WRT  weather conditios in ascending order   

```python
# plotting data groupby using seaborn

Light_vehicle = df.groupby(['Light_Conditions'], as_index=False)['Number_of_Vehicles'].sum().sort_values(by='Number_of_Vehicles', ascending=True).head()

sns.set(rc={'figure.figsize':(12,5)})
ax=sns.barplot(data = Light_vehicle, x = 'Light_Conditions',y= 'Number_of_Vehicles')
for bars in ax.containers:
    ax.bar_label(bars)
```

* Represents the number of accidents happened WRT Light_Conditions in Asc_Order

```python
# plotting data hue

plt.figure(figsize=(10,5))
ax=sns.countplot(data=df,x='Accident_Severity',hue='Urban_or_Rural_Area')
for bars in ax.containers:
    ax.bar_label(bars)
```

* Represents Accident_Severity WRT Area i.e., Rural or Urban.
